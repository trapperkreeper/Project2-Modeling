```python
# Import dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from imblearn.under_sampling import RandomUnderSampler
import xgboost as xgb
```

```python
# Reading in data
csv_path = "data/diabetes_binary_health_indicators_BRFSS2015.csv"
data = pd.read_csv(csv_path)
data.head()
```

```python
X_base = data.drop(columns='Diabetes_binary')
y_base = data['Diabetes_binary']
X_train, X_test, y_train, y_test = train_test_split(X_base, y_base, random_state=42)
model = xgb.XGBClassifier()
model.fit(X_train, y_train)

pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

print(f"Train/Test Score: {accuracy_score(y_train, pred_train)} / {accuracy_score(y_test, pred_test)}")
print(classification_report(y_test, pred_test, target_names=["negative", "positive"]))
```

```python
# X = data.drop(columns=["Diabetes_binary", "Education", "CholCheck", "AnyHealthcare", "BMI"])
# y = data["Diabetes_binary"]
# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Instantiate a RandomUnderSampler instance
rus = RandomUnderSampler(random_state=42)

# Fit the training data to the random undersampler model
X_undersampled, y_undersampled = rus.fit_resample(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid
param_grid = {
    'max_depth': [2, 3, 4],
    'learning_rate': [1, 0.1, 0.01],
    'subsample': [0.5, 0.7, 1],
    'n_estimators': [400]
}

# Create the XGBoost model object
xgb_under_grid = xgb.XGBClassifier()

# Create the GridSearchCV object
grid_search_under = GridSearchCV(xgb_under_grid, param_grid, cv=5, scoring='accuracy')

# Fit the GridSearchCV object to the training data
grid_search_under.fit(X_undersampled, y_undersampled)

grid_pred_train = grid_search_under.predict(X_undersampled)

# Make predictions on test data
grid_pred_under = grid_search_under.predict(X_test)

# Print the best set of hyperparameters and the corresponding score
print("Best set of hyperparameters: ", grid_search_under.best_params_)
print("Best score: ", grid_search_under.best_score_)
print("------------------------------------------------------")
print(f"Classification Report - XGBoost GridSearch Undersampled")
print(classification_report(y_test, grid_pred_under, labels=[1,0]))

print(f"Train Accuracy: {accuracy_score(y_undersampled, grid_pred_train)}")
print(f"Test Accuracy: {accuracy_score(y_test, grid_pred_under)}")
```

```python
lrm = LogisticRegression(penalty=None, C=0.1, max_iter=50, solver='sag', tol=0.01, random_state=42)
lrm.fit(X_undersampled, y_undersampled)

pred_train = lrm.predict(X_undersampled)
pred_test = lrm.predict(X_test)

print(f"Train/Test Score: {accuracy_score(y_undersampled, pred_train)} / {accuracy_score(y_test, pred_test)}")
print(classification_report(y_test, pred_test, target_names=["negative", "positive"]))
```

```python
rfm = RandomForestClassifier()
rfm.fit(X_undersampled, y_undersampled)
rfm_train = rfm.predict(X_undersampled)
rfm_test = rfm.predict(X_test)

print(f"Train/Test Score: {accuracy_score(y_undersampled, rfm_train)} / {accuracy_score(y_test, rfm_test)}")
print(classification_report(y_test, rfm_test, target_names=["negative", "positive"]))
```
